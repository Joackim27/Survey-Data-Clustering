Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

Read the raw dataset

```python
df = pd.read_csv('raw_data/karinderya_data.csv')
df.head()
```

Process the feature 'main_ingredients'
- check the frequency

```python
# Create a list to hold all ingredients
all_ingredients = []

# Loop through the 'main_ingredients' column
for ingredients in df['main_ingredients']:
    all_ingredients.extend(ingredients.split(','))

# Count the frequency of each ingredient
ingredient_count = Counter(all_ingredients)

# Extract the keys and values from the Counter
ingredients = list(ingredient_count.keys())
counts = list(ingredient_count.values())

plt.figure(figsize=(15, 6))  # width, height in inches
plt.bar(ingredients,counts)
plt.title("Main Ingredients")
plt.xticks(rotation=45)
plt.xlabel("Ingredients")
plt.ylabel("Frequency")
# Adjust layout to fit the labels
plt.tight_layout()

plt.show()

```

Process 'kitchen_appliances'
- used to assess whether they use uncommon appliances
- no appliance = total_respondents - owned

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# Ensure non-string values are handled
df['kitchen_appliances'] = df['kitchen_appliances'].fillna('').str.strip()

# Collect all unique kitchen appliances (removing any accidental blanks)
all_possible_appliances = set()
for kitchen_appliances in df['kitchen_appliances']:
    if isinstance(kitchen_appliances, str):
        appliances = [x.strip() for x in kitchen_appliances.split(',') if x.strip()]  # Remove extra spaces
        all_possible_appliances.update(appliances)

# Count the frequency of appliances owned
all_kitchen_appliances = []
for kitchen_appliances in df['kitchen_appliances']:
    if isinstance(kitchen_appliances, str):
        appliances = [x.strip() for x in kitchen_appliances.split(',') if x.strip()]
        all_kitchen_appliances.extend(appliances)

kitchen_appliances_count = Counter(all_kitchen_appliances)
print(kitchen_appliances_count)

# Get the total number of respondents
total_respondents = len(df)

# Compute ownership and non-ownership
appliance_owning = []
appliance_non_owning = []
appliance_labels = []

for appliance in sorted(all_possible_appliances):  # Sorting ensures consistency
    count_owning = kitchen_appliances_count.get(appliance, 0)
    count_not_owning = total_respondents - count_owning

    appliance_labels.append(appliance)
    appliance_owning.append(count_owning)
    appliance_non_owning.append(count_not_owning)


# Plot stacked bar chart
#plt.figure(figsize=(12, 6))
plt.bar(appliance_labels, appliance_owning, label="Own", color="tab:blue")
plt.bar(appliance_labels, appliance_non_owning, bottom=appliance_owning, label="Do Not Own", color="tab:orange")

# Add labels and title
plt.xlabel("Kitchen Appliance")
plt.ylabel("Number of People")
plt.title("Ownership vs. Non-Ownership of Kitchen Appliances")
plt.xticks(rotation=45, ha="right")  # Align text properly
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
